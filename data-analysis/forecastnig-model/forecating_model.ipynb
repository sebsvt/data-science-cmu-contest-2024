# Top 10 most sended product items analysis

```python
df_demand['bill_date'] = pd.to_datetime(df_demand['bill_date'])
# create a month
df_demand['month'] = df_demand['bill_date'].dt.to_period('M')
# groupby month and item and sum the quantity
monthly_demand = df_demand.groupby(["month", 'item_name'])['quantity'].sum().reset_index()
```

```python
monthly_demand
```

```python
# Get the top 10 items by demand for each month
top_items_per_month = monthly_demand.groupby('month').apply(
    lambda x: x.sort_values(by='quantity', ascending=False).head(10)
).reset_index(drop=True)
```
